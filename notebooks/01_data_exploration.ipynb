```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns


%matplotlib inline
```

```python
df=pd.read_csv("../data/AQI_Data.csv")
df.head()
```

```python
print(f"This is shape of the Data: {df.shape}")
df.info()
print("This describes the DataFrame: ")
print(f"{df.describe()}")
df.isnull().sum()
```

```python
missing_percent= df.isnull().mean()*100
print(missing_percent)
```

```python
si_cols = [col for col in df.columns if col.endswith('_SI')]
df = df.drop(columns=si_cols)
if 'AQI' in df.columns:
    df = df.drop(columns=['AQI'])
```

```python
print("Remaining columns:")
print(df.columns)
```

```python
df = df.sort_values(['Station_Id', 'Year', 'Month', 'Day', 'Hour'])
df = df.reset_index(drop=True)
```

```python
df[['CO (mg/m3)', 'CO (ug/m3)']].head()
```

```python
df[['CO (mg/m3)', 'CO (ug/m3)']].corr()
```

```python
df = df.drop(columns=['CO (mg/m3)'])
```

```python
pollutant_cols = [
    'PM2.5 (ug/m3)',
    'PM10 (ug/m3)',
    'NO2 (ug/m3)',
    'NH3 (ug/m3)',
    'SO2 (ug/m3)',
    'CO (ug/m3)',  
    'Ozone (ug/m3)'
]

df[pollutant_cols] = (
    df.groupby('Station_Id')[pollutant_cols]
      .transform(lambda x: x.interpolate(method='linear', limit_direction='both'))
)

print(df[pollutant_cols].isnull().sum())
```

```python
df.groupby('Station_Id')[['PM10 (ug/m3)', 'NH3 (ug/m3)']].apply(lambda x: x.isnull().all())
```

```python
df = df.drop(columns=['PM10 (ug/m3)', 'NH3 (ug/m3)'])
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
df['Datetime'] = pd.to_datetime(
    df['From Date'],
    format='%d-%m-%Y %H:%M'
)
```

```python
df = df.sort_values(['Station_Id', 'Datetime'])
df = df.reset_index(drop=True)
df = df.drop(columns=['Year', 'Month', 'Day', 'Hour'])
```

```python
print(df['Datetime'].min(), df['Datetime'].max())
df['Datetime'].isnull().sum()
```
