```python

stations = pd.read_csv(r'C:\Users\GARV VERMA\Desktop\Storage\codes\Projects\Pollution_Project\data\stations_info.csv')
pollution_states = set(stations['state'].unique())

disease = pd.read_csv(r'C:\Users\GARV VERMA\Desktop\Storage\codes\Projects\Pollution_Project\data\AirPollutionDeath\airpollutiondiseases1st.csv')
disease_states = set(disease['location_name'].unique())

disease_states.discard('India')

print('STATES IN POLLUTION DATA:')
print(sorted(pollution_states))
print(f'\nTotal: {len(pollution_states)}')

print('\nSTATES IN DISEASE DATA:')
print(sorted(disease_states))
print(f'\nTotal: {len(disease_states)}')

print('\nIN POLLUTION BUT NOT IN DISEASE DATA:')
print(sorted(pollution_states - disease_states))

print('\nIN DISEASE BUT NOT IN POLLUTION DATA:')
print(sorted(disease_states - pollution_states))
```
