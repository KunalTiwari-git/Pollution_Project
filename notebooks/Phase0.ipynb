Classifying Columns

```python
METADATA_COLS=[
  'From Date',
  'To Date'
]

AIR_POLLUTION_COLS=[
  'PM2.5 (ug/m3)', # Fine Particulate Matter
  'PM10 (ug/m3)',  # Coarse Particulate Matter
  'NO (ug/m3)',    #Nitric Oxcide
  'NOX (ppb)',
  'NO2 (ug/m3)',      
  'SO2 (ug/m3)',   # Causes Acid Rain
  'CO (ug/m3)',    # Carbon monoxide
  'NH3 (ug/m3)',   # Ammonia
  'Ozone (ug/m3)', # Ground Level Oxygen
]

VOC_COLS = [
    'Benzene (ug/m3)',       # Known carcinogen 
    'Toluene (ug/m3)',       # Neurotoxin
    'Eth-Benzene (ug/m3)',   # Ethylbenzene — possible carcinogen
    'MP-Xylene (ug/m3)',     # Meta+Para Xylene
    'O Xylene (ug/m3)',      # Ortho-Xylene
    'Xylene (ug/m3)',        # General xylene
]

WEATHER_COLS = [
    'WS (m/s)',         # Wind Speed — higher wind = more dispersion = lower AQI
    'WD (degree)',      # Wind Direction
    'RH (%)',           # Relative Humidity — affects particle formation
    'AT (degree C)',    # Ambient Temperature
    'SR (W/mt2)',       # Solar Radiation — affects ozone formation
    'RF (mm)',          # Rainfall — washes out particles
    'BP (mmHg)',        # Barometric Pressure
]

MINOR_COLS = [
    'CH4 (ug/m3)',      # Methane — greenhouse gas
    'NMHC (ug/m3)',     # Non-methane hydrocarbons
    'THC (ug/m3)',      # Total hydrocarbons
    'HCHO (ug/m3)',     # Formaldehyde (only 1 file)
    'Hg (ug/m3)',       # Mercury (only 1 file )
    'CO2 (mg/m3)',      # Carbon dioxide
    'SPM (ug/m3)',      # Suspended Particulate Matter 
]
```

```python
DEDUPLICATE_MAP = {

    #  NOx variants 
    'NOx (ug/m3)' : 'NOx (ppb)',
    'NOx (ppm)'   : 'NOx (ppb)',

    #  CO variants 
    'CO (ug/m3)'   : 'CO (mg/m3)',
    'CO (mg/Nm3)'  : 'CO (mg/m3)',
    'CO (ng/m3)'   : 'CO (mg/m3)',

    #  NO variants 
    'NO (mg/m3)'   : 'NO (ug/m3)',
    'NO (ppb)'     : 'NO (ug/m3)',
    'NO (ppm)'     : 'NO (ug/m3)',
    'NO ()'        : 'NO (ug/m3)',

    #  Ozone variants 
    'Ozone ()'     : 'Ozone (ug/m3)',
    'Ozone (ppb)'  : 'Ozone (ug/m3)',

    # NH3 variants 
    'NH3 ()'       : 'NH3 (ug/m3)',
    'NH3 (ppb)'    : 'NH3 (ug/m3)',

    # SO2 variants 
    'SO2 ()'       : 'SO2 (ug/m3)',

    # Temperature variants 
    'AT ()'        : 'AT (degree C)',
    'AT (degree)'  : 'AT (degree C)',
    'AT (ug/m3)'   : 'AT (degree C)',  
    'Temp (degree C)': 'AT (degree C)',
    'Temp ()'      : 'AT (degree C)',
    'Temp (ug/m3)' : 'AT (degree C)',   

    #  Relative Humidity variants 
    'RH ()'        : 'RH (%)',
    'RH (W/mt2)'   : 'RH (%)',         
    'RH (degree)'  : 'RH (%)',         

    #  Wind Speed variants 
    'WS ()'        : 'WS (m/s)',
    'WS (ug/m3)'   : 'WS (m/s)',       
    'VWS (m/s)'    : 'WS (m/s)',       

    #  Wind Direction variants 
    'WD (deg)'     : 'WD (degree)',
    'WD (degree C)': 'WD (degree)',     
    'WD ()'        : 'WD (degree)',

    #  Solar Radiation variants 
    'SR ()'        : 'SR (W/mt2)',
    'SR (ug/m3)'   : 'SR (W/mt2)',     

    #  Rainfall variants  
    'RF ()'        : 'RF (mm)',
    'RF (m/s)'     : 'RF (mm)',       
    'RF (mm).1'    : 'RF (mm)',
    'RF (mm).2'    : 'RF (mm)',
    'RF (mm).3'    : 'RF (mm)',
    'RF (mm).4'    : 'RF (mm)',
    'RF (mm).5'    : 'RF (mm)',
    'RF (mm).6'    : 'RF (mm)',
    'RF (mm).7'    : 'RF (mm)',

    #  Barometric Pressure variants
    'BP ()'        : 'BP (mmHg)',
    'BP (mg/m3)'   : 'BP (mmHg)',       
    'BP (W/mt2)'   : 'BP (mmHg)',      
    'BP (mg/m3)'   : 'BP (mmHg)',
    'BP (mmHg).1'  : 'BP (mmHg)',
    'BP (mmHg).2'  : 'BP (mmHg)',
    'BP (mmHg).3'  : 'BP (mmHg)',
    'BP (mmHg).4'  : 'BP (mmHg)',
    'BP (mmHg).5'  : 'BP (mmHg)',
    'BP (mmHg).6'  : 'BP (mmHg)',
    'BP (mmHg).7'  : 'BP (mmHg)',

    # VOC variants
    'Benzene ()'      : 'Benzene (ug/m3)',
    'Benzene (mg/m3)' : 'Benzene (ug/m3)',
    'Toluene ()'      : 'Toluene (ug/m3)',
    'Eth-Benzene ()'  : 'Eth-Benzene (ug/m3)',
    'MP-Xylene ()'    : 'MP-Xylene (ug/m3)',
    'Xylene ()'       : 'Xylene (ug/m3)',

    # CH4 variants 
    'CH4 ()'          : 'CH4 (ug/m3)',

    #  NMHC / THC variants
    'NMHC ()'         : 'NMHC (ug/m3)',
    'THC ()'          : 'THC (ug/m3)',
}
```

```python
DROP_COLS = [
    'Variance (n)',  
    'Power (W)',        # Station power reading, not environmental
    'MH (m)',          
    'VWS (degree)',     # Only 1 file
    'Gust (km/hr)',     # (4 files)
    'Gust (kl/h)',     # (2 files)
    'Gust (m/s)' ,      # (2 files)
]
```

```python
all_col_names = set(unique_columns.keys())

total_air = len(AIR_POLLUTION_COLS)
total_voc  = len(VOC_COLS)
total_wx   = len(WEATHER_COLS)
total_minor = len(MINOR_COLS)
total_dedup = len(DEDUPLICATE_MAP)
total_drop  = len(DROP_COLS)
```

COLUMN CLASSIFICATION COMPLETE

```python
print(f' Core air pollutants    :{total_air} columns')
print(f' VOCs (carcinogens)     :{total_voc} columns')
print(f' Weather / meteo        :{total_wx} columns')
print(f' Minor / rare           :{total_minor} columns')
print(f' Duplicates to merge    :{total_dedup} messy -> clean mappings')
print(f' Columns to drop        :{total_drop} columns')
```

```python
all_classified = (
    set(METADATA_COLS) |
    set(AIR_POLLUTION_COLS) |
    set(VOC_COLS) |
    set(WEATHER_COLS) |
    set(MINOR_COLS) |
    set(DEDUPLICATE_MAP.keys()) |
    set(DEDUPLICATE_MAP.values()) |
    set(DROP_COLS)
)

truly_unclassified = [c for c in all_col_names if c not in all_classified]

if truly_unclassified:
    print('UNCLASSIFIED — review these:')
    for col in truly_unclassified:
        print(f'   ? {col}  ({unique_columns[col]} files)')
else:
    print('Every column is classified')
```

```python
rows = []
for col, count in unique_columns.items():
    if col in AIR_POLLUTION_COLS:       bucket = 'air_pollution'
    elif col in VOC_COLS:               bucket = 'voc'
    elif col in WEATHER_COLS:           bucket = 'weather'
    elif col in MINOR_COLS:             bucket = 'minor'
    elif col in METADATA_COLS:          bucket = 'metadata'
    elif col in DEDUPLICATE_MAP:        bucket = f'duplicate_of → {DEDUPLICATE_MAP[col]}'
    elif col in DROP_COLS:              bucket = 'drop'
    else:                               bucket = 'unclassified'

    rows.append({
        'column_name'   : col,
        'files_count'   : count,
        'coverage_pct'  : round((count / len(station_data_files)) * 100, 1),
        'bucket'        : bucket
    })
```

CHECKING DATA RANGE 

```python
date_sample = random.sample(station_data_files, min(20, len(station_data_files)))
all_min_years = []
all_max_years = []

for file_path in date_sample:
    filename = os.path.basename(file_path)
    try:
        df = pd.read_csv(file_path, usecols=['From Date'])
        dates = pd.to_datetime(df['From Date'], errors='coerce').dropna()
        
        if len(dates) > 0:
            min_year = dates.dt.year.min()
            max_year = dates.dt.year.max()
            
            span = max_year - min_year + 1
            
            all_min_years.append(min_year)
            all_max_years.append(max_year)
            
            print(f'  {filename:<15} -> {min_year} to {max_year}  ({span} years)')
        else:
            print(f'{filename} No valid dates found')
    except Exception as e:
        print(f'{filename} Error: {e}')
```

```python
if all_min_years:
    print()
    print('OVERALL DATASET DATE RANGE:')
    print(f'Earliest data from : {min(all_min_years)}')
    print(f'Latest data from   : {max(all_max_years)}')
    print(f'Max possible span  : {max(all_max_years) - min(all_min_years) + 1} years')
```
