IMPORTING LIBRARIES

```python
import pandas as pd
import os
import glob
import random
```

SET PATHS

```python
DATA_DIR= r'C:\Users\GARV VERMA\Desktop\Storage\codes\Projects\Pollution_Project\data'
OUTPUT_DIR = r'C:\Users\GARV VERMA\Desktop\Storage\codes\Projects\Pollution_Project\outputs'

if os.path.exists(DATA_DIR):
    print(f'Data folder found: {os.path.abspath(DATA_DIR)}')
else:
    print(f'ERROR: Data folder not found')
    print('Please update  DATA_DIR above to correct path!')

os.makedirs(OUTPUT_DIR,exist_ok=True)
print(f'Output folder ready: {os.path.abspath(OUTPUT_DIR)}')
```

COUNT AND LIST ALL FILES

```python
all_csv_files= glob.glob(os.path.join(DATA_DIR,'*.csv'))

stations_info_file=None
station_data_files=[]

for file in all_csv_files:
    filename = os.path.basename(file)
    if filename.lower() == 'stations_info.csv':
        stations_info_file = file
    else:
        station_data_files.append(file)
```

```python
print(f'Total CSV files found: {len(all_csv_files)}')
print(f' Stations_info.csv:{"Found" if stations_info_file else "NOT FOUND"}')
print(f' Station data files: {len(station_data_files)}')
print()

print('Sample file names: ')
for f in sorted(station_data_files)[:10]:
    print(f'')
```
